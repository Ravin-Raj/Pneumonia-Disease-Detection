Count the Dataset

```python
import os

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Base path to your dataset
base_path = '/content/drive/MyDrive/Pneumonia Disease Detection Dataset'

# Dataset types and classes
data_types = ['Training Data', 'Validation Data', 'Testing Data']
classes = ['Normal', 'Bacterial Pneumonia', 'Viral Pneumonia']

grand_total = 0  # For entire dataset

# Count and print
for dtype in data_types:
    print(f'\nğŸ“ {dtype}')
    type_total = 0  # Total images for this dataset type
    for cls in classes:
        folder_path = os.path.join(base_path, dtype, cls)
        if os.path.exists(folder_path):
            count = len([
                f for f in os.listdir(folder_path)
                if f.lower().endswith(('.jpg', '.jpeg', '.png'))
            ])
            print(f' - {cls}: {count} images')
            type_total += count
        else:
            print(f' - {cls}: âŒ Folder not found')

    print(f'âœ… Total in {dtype}: {type_total} images')
    grand_total += type_total

print(f'\nğŸ§® Grand Total Images in Dataset: {grand_total} images')
```
