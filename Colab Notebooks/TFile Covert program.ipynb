```python
# Set your model paths
h5_model_path = "/content/drive/MyDrive/Best MobileNetV2 Trained Model.h5"
tflite_model_path = "/content/drive/MyDrive/Best MobileNetV2 Trained Model.tflite"
```

```python
import tensorflow as tf

# Load Keras model
model = tf.keras.models.load_model(h5_model_path)

# Convert to TFLite
converter = tf.lite.TFLiteConverter.from_keras_model(model)

# Optional: optimization for mobile
converter.optimizations = [tf.lite.Optimize.DEFAULT]

# Convert the model
tflite_model = converter.convert()

# Save the .tflite model back to Google Drive
with open(tflite_model_path, "wb") as f:
    f.write(tflite_model)

print("✅ TFLite model saved at:", tflite_model_path)
```
