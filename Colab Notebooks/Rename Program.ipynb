Rename the Dataset

```python
import os

from google.colab import drive
drive.mount('/content/drive')

# Base dataset path
base_path = '/content/drive/MyDrive/Pneumonia Disease Detection Dataset'

# Dataset and class folder names
data_types = ['Training Data', 'Validation Data', 'Testing Data']
classes = ['Normal', 'Bacterial Pneumonia', 'Viral Pneumonia']

# Rename images with format: Training_Data_Normal_0001.jpg
for dtype in data_types:
    dtype_clean = dtype.replace(" ", "_")  # e.g., Training_Data
    print(f'\nğŸ“ Renaming in: {dtype}')

    for cls in classes:
        cls_clean = cls.replace(" ", "_")  # e.g., Bacterial_Pneumonia
        folder_path = os.path.join(base_path, dtype, cls)

        if not os.path.exists(folder_path):
            print(f' - âŒ Missing folder: {folder_path}')
            continue

        files = sorted(os.listdir(folder_path))
        count = 0

        for i, filename in enumerate(files):
            ext = os.path.splitext(filename)[-1].lower()
            if ext not in ['.jpg', '.jpeg', '.png']:
                continue

            new_filename = f"{dtype_clean}_{cls_clean}_{i+1:04d}{ext}"
            src = os.path.join(folder_path, filename)
            dst = os.path.join(folder_path, new_filename)

            os.rename(src, dst)
            count += 1

        print(f' - âœ… Renamed {count} images in {cls}')
```
